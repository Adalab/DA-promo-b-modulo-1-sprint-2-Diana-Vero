```python
#Hagamos unas queries facilitas:
#Extraed el primer pedido donde se ha gastado más. Devolved el id del pedido y la cantidad gastada.

import mysql.connector

mydb = mysql.connector.connect(
                  user='root', password='AlumnaAdalab',
                  host='127.0.0.1',
                  database='northwind')

mycursor = mydb.cursor()

query= 'SELECT order_id, (unit_price*quantity) AS cantidad_gastada FROM order_details ORDER BY cantidad_gastada DESC LIMIT 1;'

mycursor.execute(query)

for x in mycursor:
  print(x)

mydb.close()
```

```python
#Extraer los pedidos hechos por "Nancy Daviolo". Tendréis que hacer una subquery.

import mysql.connector

mydb = mysql.connector.connect(
                  user='root', password='AlumnaAdalab',
                  host='127.0.0.1',
                  database='northwind')
import pandas as pd
query= '''SELECT employees.first_name, employees.last_name, orders.order_id
FROM employees
INNER JOIN orders
ON employees.employee_id = orders.employee_id
WHERE employees.employee_id = 1
GROUP BY orders.order_id;'''

df= pd.read_sql_query(query, mydb)
display(df)   #NO CONSIGO AGRUPAR EL NÚMERO DE PEDIDOS

mydb.close()
```
